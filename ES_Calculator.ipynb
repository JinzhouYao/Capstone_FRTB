```python
data = pd.DataFrame(dict1)
data
```

```python
def ES(PL: str) -> float:
    Data = pd.read_excel(PL)
    L1=[] #IR pvList
    L2=[] #FX pvList
    L3=[] #EQ pvList
    L4=[] #TOTAL pvList
    for i in np.linspace(11,209,19):
        i=int(i)
        L1.append(list(map(float,Data.iloc[i+7,23].split(' '))))
        L2.append(list(map(float,Data.iloc[i+8,23].split(' '))))
        L3.append(list(map(float,Data.iloc[i+9,23].split(' '))))
        L4.append(list(map(float,Data.iloc[i+10,23].split(' '))))
    dict1={'IR':L1,'FX':L2,'EQ':L3,'TOTAL':L4}
    data = pd.DataFrame(dict1)
    PLL=[[],[],[],[]]
    k=0
    for j in data.columns.tolist():
        for i in data.index:
            v1=data.at[i,j]
            v2=v1.copy()
            v2.insert(0,v1[0])
            v = list(map(lambda x: x[0]-x[1], zip(v1, v2)))
            del v[0]
            PLL[k].append(v)
        k+=1
    dict2={'IR':PLL[0],'FX':PLL[1],'EQ':PLL[2],'TOTAL':PLL[3]}
    P_L = pd.DataFrame(dict2)
    ESL=[[],[],[],[]]
    k=0
    for j in data.columns.tolist():
        for i in data.index:
            t1=P_L.at[i,j]
            t2=sorted(t1)
            cvar=t2[0]+t2[1] 
            ESL[k].append(cvar)
        k+=1
    dict3={'ES_IR':ESL[0],'ES_FX':ESL[1],'ES_EQ':ESL[2],'ES_TOTAL':ESL[3]}
    ES = pd.DataFrame(dict3)
    ES.to_excel("ES.xlsx")
ES("HVAR_10D.xlsx")
```
